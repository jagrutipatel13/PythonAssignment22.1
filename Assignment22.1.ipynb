```python
import numpy as np
from sklearn import cluster, datasets
from scipy import misc
import scipy as sp
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans

```

```python
face = sp.misc.face(gray = True)
n_clusters = 5
np.random.seed(0)
X = face.reshape((-1,1))
k_Means= cluster.KMeans(n_clusters=n_clusters,n_init=4)
k_Means.fit(X)
plt.subplot(131)
plt.imshow(face, cmap=plt.cm.gray)

plt.subplot(132)
plt.imshow(face, cmap=plt.cm.gray, vmin=30, vmax=200)
plt.axis('off')

plt.subplot(133)
plt.imshow(face, cmap=plt.cm.gray)
plt.contour(face, [50, 200])
plt.axis('off')

plt.subplots_adjust(wspace=0, hspace=0., top=0.99, bottom=0.01, left=0.05,
                    right=0.99)
plt.show()

```
